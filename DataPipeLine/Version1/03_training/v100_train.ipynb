이 노트북은 V100 GPU 서버 환경에서 영양제 검출 모델을 학습시키기 위해 작성되었습니다.

```python
# 2. 데이터셋 경로 설정 확인
# 주피터 서버의 현재 경로에 맞게 data.yaml을 수정해야 할 수 있습니다.
import yaml
import os

yaml_path = 'datasets/supplements/data.yaml'

with open(yaml_path, 'r') as f:
    data = yaml.safe_load(f)

# 서버 환경에 맞게 경로를 상대 경로로 변경
data['path'] = os.path.abspath('datasets/supplements')
data['train'] = 'images/train'
data['val'] = 'images/val'

with open(yaml_path, 'w') as f:
    yaml.dump(data, f)

print("✅ data.yaml 경로 최적화 완료:")
print(data)
```

```python
# 3. 학습 시작 (V100 특화 설정)
from ultralytics import YOLO

# 모델 로드 (yolo11m 또는 yolo11l 추천)
model = YOLO('yolo11m.pt')

# 학습 실행
results = model.train(
    data='datasets/supplements/data.yaml',
    epochs=100,         # V100이면 100~200 에포크 권장
    imgsz=640,          # 또는 1280 (고해상도 학습 가능)
    batch=32,           # V100 메모리가 넉넉하므로 32 이상 가능
    device=0,           # GPU 사용
    project='runs/train',
    name='supplements_v100_v1',
    exist_ok=True
)
```

```python
# 4. 학습 결과 모델 다운로드 안내
print("🏆 학습 완료!")
print("최종 모델 위치: runs/train/supplements_v100_v1/weights/best.pt")
# 이 파일을 다운로드해서 로컬에서 테스트하면 됩니다.
```

## 📊 Final Performance Comparison Table
마지막으로 각 버전의 핵심 지표를 비교합니다.

```python
import matplotlib.pyplot as plt

def get_best_metric(dir_path, metric_col):
    csv_path = os.path.join(dir_path, 'results.csv')
    if os.path.exists(csv_path):
        df = pd.read_csv(csv_path)
        df.columns = [c.strip() for c in df.columns]
        return df[metric_col].max()
    return 0.0

metrics = ['metrics/mAP50(B)', 'metrics/precision(B)', 'metrics/recall(B)']
v1_scores = [get_best_metric(DIRS['V1 (Baseline)'], m) for m in metrics]
v2_scores = [get_best_metric(DIRS['V2 (Augmented)'], m) for m in metrics]
v3_scores = [get_best_metric(DIRS['V3 (Final)'], m) for m in metrics]

df_comp = pd.DataFrame({
    'Metric': ['mAP@50', 'Precision', 'Recall'],
    'V1': v1_scores,
    'V2': v2_scores,
    'V3': v3_scores
})

display(df_comp)

df_comp.set_index('Metric').plot(kind='bar', figsize=(12, 6), rot=0, colormap='viridis')
plt.title('Performance Comparison across Versions', fontsize=16)
plt.ylabel('Score')
plt.ylim(0, 1.05)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.legend(title='Version')
plt.show()
```
